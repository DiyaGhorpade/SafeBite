```python
#convert each column to string,drop all missing records and display unique entries in each column
df['CountryName'] = df['CountryName'].astype(str)
df['FoodName'] = df['FoodName'].astype(str)
df['FoodGroupName'] = df['FoodGroupName'].astype(str)
df['ContaminantName'] = df['ContaminantName'].astype(str)
unique_countries = df['CountryName'].dropna().unique()
unique_foodname = df['FoodName'].dropna().unique()
unique_foodgroup = df['FoodGroupName'].dropna().unique()
unique_contaminantname = df['ContaminantName'].dropna().unique()
print("Unique Countries:", unique_countries)
print("Unique Food Names:", unique_foodname)
print("Unique Food Groups:", unique_foodgroup)
print("Unique Contaminant Names:", unique_contaminantname)
```

```python
#remove resultvalue with 0
df = df[df['ResultValue'].notna() & (df['ResultValue'] != 0)]
print((df['ResultValue'] == 0).sum())
```

```python
#Find correlation between foodname and gemsfoodname
common_values = df['FoodName'].isin(df['GEMSFoodName']).sum()
print(f"Number of common entries between FoodName and GEMSFoodName: {common_values}")
```

```python
#Cramer's V Correlation to drop categorical columns that are highly correlated
import scipy.stats as stats

# Create a contingency table
contingency_table = pd.crosstab(df['FoodName'], df['GEMSFoodName'])

# Compute chi-squared statistic
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Compute Cramér's V
n = contingency_table.sum().sum()  # Total observations
phi2 = chi2 / n
r, k = contingency_table.shape
cramers_v = np.sqrt(phi2 / min(r - 1, k - 1))

print(f"Cramér's V correlation: {cramers_v:.4f}")
```

```python
#Drop irrelevant columns
df = df.drop(columns=['FoodName','FoodGroupID', 'FoodEx2Code', 
                      'CountryID','UnitName', 'UnitID','WHO_Food_Code','RegionCode','SerialNumber'])

print(df.columns) 
```

```python
#Focus on Asian region & Indian Subcontinent-Remove Australia & NewZealand
df = df[~df['CountryName'].isin(['Australia', 'New Zealand'])]
print(df['CountryName'].unique()) 
```

```python
non_perishable_groups = [
    "Alcoholic beverages",
    "Stimulant beverages, dried and diluted excluding cocoa products",
    "Composite food (including frozen products)",
    "Other foods",
    "Food for infants and small children",
    "Sugar and confectionary (including cocoa products)",
    "Non-alcoholic beverages (excluding milk, fruit and vegetable juice, water and stimulants)",
    "Drinking water (water without any additives except carbon dioxide; includes water ice for consumption)",
    "Fruit and vegetable juices",
    "Fats and oils of animal and vegetable (excluding butter)",
    "Snacks and desserts",
    "Products for special nutritional use"
]
df=df[~df['FoodGroupName'].isin(non_perishable_groups)]
print(df['FoodGroupName'].unique())
```

```python
df.info()
```

```python
#Remove Contamination level outliers
Q1 = df['ResultValue'].quantile(0.25)
Q3 = df['ResultValue'].quantile(0.75)

# Compute Interquartile Range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['ResultValue'] < lower_bound) | (df['ResultValue'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

df = df[(df['ResultValue'] >= lower_bound) & (df['ResultValue'] <= upper_bound)]
print(df.shape)  
print(df['ResultValue'].describe())  

```

```python
df.info()
```

```python
#Check for missing records/NaN values
missing=df.isna().sum()
print(missing)
```

```python
#check for number of unique entries
for column in df.columns:
    unique_values = df[column].nunique()
    print(f"{column}: {unique_values}")
```

```python
#vIsualize top 20 contaminants in a bar chart
top_contaminants = df['ContaminantName'].value_counts().head(20)

plt.figure(figsize=(10,5))
sns.barplot(x=top_contaminants.values, y=top_contaminants.index, palette="magma")
plt.xlabel("Number of Records")
plt.ylabel("Contaminant Name")
plt.title("Top 20 Most Reported Contaminants in Food")
plt.show()
```

```python
#Visualize contaminant levels v/s contaminant name
contaminant_avg = df.groupby('ContaminantName')['ResultValue'].mean().reset_index()
top_contaminants = contaminant_avg.head(20)

plt.figure(figsize=(12,6))
sns.barplot(x='ResultValue', y='ContaminantName', hue='ContaminantName',data=top_contaminants, palette="Reds_r")
plt.xlabel("Average Contamination Level")
plt.ylabel("Contaminant Name")
plt.title("Top 20 Contaminants with Highest Contamination Levels")
plt.show()
```
